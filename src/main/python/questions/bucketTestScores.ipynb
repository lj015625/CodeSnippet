### Let’s say you’re given a dataframe of standardized test scores from high schoolers from grades 9 to 12 called df_grades. Given the dataset, write code function in Pandas called bucket_test_scores to return the cumulative percentage of students that received scores within the buckets of <50, <75, <90, <100.

```python
import pandas as pd
```

```python
df_grades = pd.DataFrame([[1, 10, 85], [2, 10, 60], [3, 11, 90], [4, 10, 30], [5, 11, 99]],
                         columns=['id', 'grade', 'score'])
```

```python
print(df_grades['grade'].unique())
```

```python
df_grades['score_bin'] = pd.cut(x=df_grades['score'], bins=[0, 50, 75, 90, 100],
                            labels=['<50', '<75', '<90', '<100'], right=False)
df_grades
```

```python
groupby_by_grade_score_bin = df_grades.groupby(['grade', 'score_bin']).agg({'id': 'count'}).fillna(0)
groupby_by_grade_score_bin
```

```python
groupby_by_grade_score_bin_percent = groupby_by_grade_score_bin.groupby(['grade']).apply(lambda x : 100 * x / float(x.sum()))
groupby_by_grade_score_bin_percent
```

```python
groupby_by_grade_score_bin_percent_cumsum = groupby_by_grade_score_bin_percent.groupby(['grade']).cumsum().reset_index()
groupby_by_grade_score_bin_percent_cumsum
```

```python
groupby_by_grade_score_bin_percent_cumsum['percentage'] = round(groupby_by_grade_score_bin_percent_cumsum['id'], 2).astype(str) + '%'
groupby_by_grade_score_bin_percent_cumsum.drop(columns='id', inplace=True)
groupby_by_grade_score_bin_percent_cumsum
```
