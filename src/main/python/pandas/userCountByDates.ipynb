```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
plt.style.use("fivethirtyeight")
```

```python
search_interaction_df = pd.DataFrame({"request_path" : ['search', 'search', 'search', 'search', 'search', 'search'], 
                "search_term" : ['Saving Private Ryan', 'Fear and Loathing', 'Leagally Blond', 'Leagally Blond', 'The Hill have eyes', 'Knives Out'], 
                "user_id" : ['09663ea6', '63f84e80', '31c73683', '010b4076', '31c73683','8173164f'],
                "ip_adress" : ['168.196.83.238','116.103.0.64','9.47.206.231','198.45.207.12','248.212.242.192','65.166.90.163'],
                "date" : ['2021-04-01','2021-04-01','2021-04-01','2021-04-01','2021-04-01','2021-04-01']})
search_interaction_df['date'] = pd.to_datetime(search_interaction_df['date'])
search_interaction_df['user_id'] = search_interaction_df['user_id'].astype(str)
search_interaction_df.head()
```

```python
latest_date = search_interaction_df.agg({'date': 'max'})['date']
latest_date
```

```python
one_day_before_latest_date = latest_date - timedelta(days=1)
seven_days_before_latest_date = latest_date - timedelta(days=7)
thirty_days_before_latest_date = latest_date - timedelta(days=30)
```

```python
one_day_before_latest_date_count = search_interaction_df[search_interaction_df['date'].between(one_day_before_latest_date, latest_date)].groupby('user_id').count().reset_index()
seven_days_before_latest_date_count = search_interaction_df[search_interaction_df['date'].between(seven_days_before_latest_date, latest_date)].groupby('user_id').count().reset_index()
thirty_days_before_latest_date_count = search_interaction_df[search_interaction_df['date'].between(thirty_days_before_latest_date, latest_date)].groupby('user_id').count().reset_index()
```

```python
one_day_before_latest_date_count
```

```python
one_day_before_latest_date_count = one_day_before_latest_date_count[['user_id','request_path']].rename(columns={'request_path': 'day_user_count'})
one_day_before_latest_date_count
```

```python
seven_days_before_latest_date_count = seven_days_before_latest_date_count[['user_id','request_path']].rename(columns={'request_path': 'week_user_count'})
seven_days_before_latest_date_count
```

```python
thirty_days_before_latest_date_count = thirty_days_before_latest_date_count[['user_id','request_path']].rename(columns={'request_path': 'month_user_count'})
thirty_days_before_latest_date_count
```

```python
total_counts = one_day_before_latest_date_count.merge(seven_days_before_latest_date_count, how='left', on='user_id').merge(thirty_days_before_latest_date_count, how='left', on='user_id')
total_counts = total_counts.fillna(0)
total_counts
```
